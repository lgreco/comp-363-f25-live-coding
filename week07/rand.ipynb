### Linear Congruential Generator

$$
X_{n+1} = (aX_n+c)\bmod m
$$


```python
def lcg(m, a, c, s):
    count = 0
    while True:
        s = (a*s+c) % m
        print(f"{s:12d}",end ="")
        if count % 8 == 0:
            print()
        count += 1
```

```python
(lcg(2_147_483_647, 16_807, 0, 1234))
```
